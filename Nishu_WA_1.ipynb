1. Explain the difference between list and tuple with a memory-usage example.

| List                      | Tuple                          |
| ------------------------- | ------------------------------ |
| Mutable (can be modified) | Immutable (cannot be modified) |
| Uses more memory          | Uses less memory               |
| Created using `[]`        | Created using `()`             |


```python
import sys
lst = [1, 2, 3, 4, 5]
tup = (1, 2, 3, 4, 5)
print(sys.getsizeof(lst))
print(sys.getsizeof(tup))
```

4. Write a program to check if a string is a palindrome.

```python
text = input("Enter a string: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

7. Write a script to count the number of vowels in a user-provided sentence.


```python
sentence = input("Enter a sentence: ")

count = 0

for ch in sentence.lower():
    if ch in "aeiou":
        count += 1

print("Number of vowels:", count)
```

9. Create a list of numbers from 1 to 20 and print only the even ones

```python
numbers = list(range(1, 21))

for num in numbers:
    if num % 2 == 0:
        print(num)
```
