#Question 2
 Explain the difference between integer, float and complex
data types.


| Integer (`int`)      | Float (`float`)          | Complex (`complex`)                          |
| -------------------- | ------------------------ | -------------------------------------------- |
| Stores whole numbers | Stores decimal numbers   | Stores numbers with real and imaginary parts |
| No decimal point     | Contains a decimal point | Uses `j` for imaginary part                  |
| Example: `25`        | Example: `25.5`          | Example: `3+4j`                              |


```python
a = 25
b = 25.5
c = 3 + 4j

print("Integer:", a)
print("Float:", b)
print("Complex:", c)

print(type(a))
print(type(b))
print(type(c))
```

#Question 3
What is type conversion? Explain implicit and explicit
conversion.

Type conversion is the process of converting a value from one data type to another.

1. Implicit Type Conversion

Python automatically converts one data type into another.

2. Explicit Type Conversion

The programmer converts the data type manually using functions such as int(), float(), str(), and bool().

```python


a = 10
b = 5.5

c = a + b

print(c)
print(type(c))

```

```python
a = "25"

b = int(a)

print(b)
print(type(b))
```

#Question 4
Write a program to accept two numbers and display their
sum, difference, product and quotient.

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2

print("Sum =", sum)
print("Difference =", difference)
print("Product =", product)

if num2 != 0:
    quotient = num1 / num2
    print("Quotient =", quotient)
else:
    print("Division by zero is not possible.")
```

#Question 5
Write a program to convert temperature from Celsius to
Fahrenheit.

```python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (9 / 5) * celsius + 32

print("Temperature in Fahrenheit =", fahrenheit)
```

#Question 6
Write a program to calculate the area and perimeter of a
rectangle.

```python
length = float(input("Enter length: "))
breadth = float(input("Enter breadth: "))

area = length * breadth
perimeter = 2 * (length + breadth)

print("Area =", area)
print("Perimeter =", perimeter)
```
