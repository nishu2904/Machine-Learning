# Question 1
Name the main built-in data types used in Python.


| Data Type  | Description                               | Example           |
| ---------- | ----------------------------------------- | ----------------- |

#Question 3
What is type conversion? Explain implicit and explicit
conversion.

Type conversion is the process of converting a value from one data type to another.

1. Implicit Type Conversion

Python automatically converts one data type into another.

2. Explicit Type Conversion

The programmer converts the data type manually using functions such as int(), float(), str(), and bool().

```python


a = 10
b = 5.5

c = a + b

print(c)
print(type(c))

```

#Question 4
Write a program to accept two numbers and display their
sum, difference, product and quotient.

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2

print("Sum =", sum)
print("Difference =", difference)
print("Product =", product)

if num2 != 0:
    quotient = num1 / num2
    print("Quotient =", quotient)
else:
    print("Division by zero is not possible.")
```

#Question 5
Write a program to convert temperature from Celsius to
Fahrenheit.

```python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (9 / 5) * celsius + 32

print("Temperature in Fahrenheit =", fahrenheit)
```

#Question 6
Write a program to calculate the area and perimeter of a
rectangle.

```python
length = float(input("Enter length: "))
breadth = float(input("Enter breadth: "))

area = length * breadth
perimeter = 2 * (length + breadth)

print("Area =", area)
print("Perimeter =", perimeter)
```

#Question 8
Explain arithmetic, comparison, logical and assignment
operators.

1. Arithmetic Operators

These operators perform mathematical operations.
| Operator | Description    | Example |
| -------- | -------------- | ------- |
| `+`      | Addition       | `5+3`   |
| `-`      | Subtraction    | `5-3`   |
| `*`      | Multiplication | `5*3`   |
| `/`      | Division       | `5/3`   |
| `//`     | Floor Division | `5//3`  |
| `%`      | Modulus        | `5%3`   |
| `**`     | Exponentiation | `2**3`  |
2. Comparison Operators

Used to compare two values.
| Operator | Description              |
| -------- | ------------------------ |
| `==`     | Equal to                 |
| `!=`     | Not equal to             |
| `>`      | Greater than             |
| `<`      | Less than                |
| `>=`     | Greater than or equal to |
| `<=`     | Less than or equal to    |
3. Logical Operators

Used to combine conditions.

| Operator | Description                                    |
| -------- | ---------------------------------------------- |
| `and`    | Returns True if both conditions are True       |
| `or`     | Returns True if at least one condition is True |
| `not`    | Reverses the Boolean value                     |


```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

```python
a = 10
b = 5

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
a = True
b = False

print(a and b)
print(a or b)
print(not a)
```

#Question 9
What is the difference between division and floor division?

| Division (`/`)                                        | Floor Division (`//`)                                                         |
| ----------------------------------------------------- | ----------------------------------------------------------------------------- |
| Returns the exact quotient as a floating-point number | Returns the integer quotient by removing the decimal part (floors the result) |
| Example: `10/3 = 3.3333`                              | Example: `10//3 = 3`                                                          |


```python
a = 10
b = 3

print("Division =", a / b)
print("Floor Division =", a // b)
```

#Question 10
What is operator precedence? Why is it important?



Operator precedence is the order in which Python evaluates operators in an expression. Operators with higher precedence are evaluated before operators with lower precedence. Parentheses () can be used to change the order of evaluation.

Importance
Ensures expressions are evaluated correctly.
Removes ambiguity in calculations.
Helps write accurate and readable programs.

```python
result1 = 2 + 3 * 4
result2 = (2 + 3) * 4

print("Without Parentheses:", result1)
print("With Parentheses:", result2)
```
